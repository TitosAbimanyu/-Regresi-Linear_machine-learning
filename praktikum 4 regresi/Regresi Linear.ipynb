# Import Library dan Baca Data

```python
# Import library pandas untuk manipulasi dan analisis data
import pandas as pd
# Import library numpy untuk operasi matematik
import numpy as np
# Import matplotlib.pyplot dan seaborn untuk visualisasi data
import matplotlib.pyplot as plt
import seaborn as sns
```

## Dataset1 Salary

```python
data1 = pd.read_csv('salary_dataset.csv')
data1.head()
```

```python
data1 = data1.drop('no',axis=1)
data1.head()
```

```python
plt.scatter(data1.YearsExperience, data1.Salary, color='red')
plt.xlabel('Pengalaman')
plt.ylabel('Gaji')
plt.title("Plot Data Pengalaman & Gaji")
plt.show()
```

## Dataset 2 penjualan

```python
data2 = pd.read_csv('data-penjualan.csv', delimiter=';')
data2.head()
```

```python
plt.scatter(data2.promosi, data2.penjualan, color='blue')
plt.xlabel('Promosi')
plt.ylabel('Penjualan')
plt.title("Plot Data Promosi & Penjualan")
plt.show()
```

# Korelasi

```python
data1.corr()
```

```python
data2.corr()
```

# Split Kolom

```python
#Data Salary
pengalaman = data1['YearsExperience'].values[:,np.newaxis]
gaji = data1['Salary'].values

#Data Penjualan
promosi = data2['promosi'].values[:, np.newaxis]
penjualan = data2['penjualan'].values
```

# Modelling

```python
# Import LinearRegression dari sklearn untuk membuat model regresi linear
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
model2 = LinearRegression()

model1.fit(pengalaman, gaji)
model2.fit(promosi, penjualan)
```

# Testing

```python
pengalaman_test = [[1.1],[1],[7]]
predik_gaji = model1.predict(pengalaman_test)

promosi_test = [[20],[100]]
predik_penjualan = model2.predict(promosi_test)

print("Prediksi Gaji :", predik_gaji)
print("Prediksi Penjualan :", predik_penjualan)
```

# Akurasi

```python
#Mengecek Akurasi Data
from sklearn.metrics import r2_score

# Menghitung R-squared untuk model gaji
r2_gaji = r2_score(gaji, model1.predict(pengalaman))
print("R-squared Gaji :", r2_gaji)

# Menghitung R-squared untuk model penjualan
r2_penjualan = r2_score(penjualan, model2.predict(promosi))
print("R-squared Penjualan :", r2_penjualan)

#Nilai R-squared berkisar antara 0 dan 1, dan semakin mendekati 1, semakin baik modelnya. 
#Nilai yang mendekati 0 menunjukkan bahwa model tidak dapat menjelaskan variasi data dengan baik.
```
